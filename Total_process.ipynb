## &emsp;3.2 Data Imputation Strategy

To fill missing values, an appropriate strategy was selected based on the nature of each variable:

### &emsp;&emsp;3.2.1 Preparation For Data Processing

```python
df_filled = df.copy()  # Create a copy of the dataset for processing
```

```python
# Compute the mean for each numerical column, grouped by country
country_group = df.groupby("Country").mean()

# Display the selected numerical columns for better understanding
print("ðŸ“Š Country-Wise Mean Values:")
country_group[[
    "Renewable Energy (%)", 
    "Government Investment (Million USD)", 
    "Emissions Reduction (%)",
    "Solar", "Wind", "Hydro", "Geothermal", "Biomass",
    "Other Renewables"
]]
```

```python
# Identify all numeric columns
numeric_cols = df.select_dtypes(include='number').columns.tolist()

# Define the column for government investment
gov_col = "Government Investment (Million USD)"

# Identify all numeric columns except government investment
other_num_cols = [col for col in numeric_cols if col != gov_col]
```

### &emsp;&emsp;3.2.2 Filling Missing Values In "Government Investment(Million USD)

Since government investment figures tend to have large variations and potential outliers, the median value per country was used to fill missing values:

```python
# Fill missing values in "Government Investment (Million USD)" using the median (to resist outliers)
df_filled[gov_col] = df_filled.groupby("Country")[gov_col].transform(
    lambda x: x.fillna(x.median())  
)
```

### &emsp;&emsp;3.2.3 Filling Missing Values In Other Numerical Coluns

For variables related to Renewable Energy (%) and Energy Sources (Solar, Wind, Hydro, etc.), the mean value per country was used:

```python
# Fill missing values in all other numerical columns using the mean (to preserve the central tendency)
df_filled[other_num_cols] = df_filled.groupby("Country")[other_num_cols].transform(
    lambda x: x.fillna(x.mean())  
)
```

* Using the mean ensures that missing values are filled with representative values based on each countryâ€™s historical trends.
* This method ensures continuity in visualizations.


## &emsp;3.3 Validating Missing Values after Cleaning

After handling missing values through median and mean imputation, it was essential to validate whether the missing values had been successfully addressed. This step ensured that the dataset was fully complete and ready for further analysis.

The following code was executed to check for any remaining missing values:


```python
print("ðŸ“Š Missing Values After Cleaning:")
df_filled.isnull().sum()
```

*Overview*

This visualization presents the renewable energy adoption trends for 10 major countries from 2000 to 2023. 

Each subplot represents a country, illustrating the percentage of renewable energy in total energy consumption over time. 

The goal of this analysis is to identify the patterns, trends, and key shifts in renewable energy adoption across different nations.

## &emsp;4.3  Visualization 3ï¼š
## &emsp;_Scatter Plot with Regression: Renewable Energy Share vs. Emissions Reduction_


**Overview**

This analysis explores the relationship between renewable energy adoption and emission reduction across 10 major countries.
* X-axis: Percentage of renewable energy in total energy consumption.
* Y-axis: Percentage reduction in carbon emissions.
* Blue dots represent individual data points for each year, while the red regression line (with confidence intervals) indicates the overall trend.

Additionally, a correlation analysis was performed, calculating slope, intercept, R-value (correlation coefficient), 

and p-value to determine the strength and significance of the relationship between these two variables.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy import stats

# === Data Validation ========================================================
print("\nðŸ” [3.3] Data Validation:")
print("Cleaned dataset shape:", df_filled.shape)
print("Missing values in key columns:")
print(df_filled[['Renewable Energy (%)', 'Emissions Reduction (%)']].isnull().sum())

# Filter countries with sufficient data points
country_counts = df_filled.groupby('Country')['Emissions Reduction (%)'].count()
valid_countries = country_counts[country_counts >= 5].index.tolist()
df_filtered = df_filled[df_filled['Country'].isin(valid_countries)]

print(f"\nCountries analyzed: {len(valid_countries)}")

# === Visualization Setup =====================================================
plt.rcParams.update({'font.size': 12})
sns.set_theme(style="whitegrid", palette="muted")

# === Determine Layout for Subplots ===========================================
cols = 4  # Number of columns in the grid
rows = int(np.ceil(len(valid_countries) / cols))  # Dynamically calculate rows

fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(16, 4 * rows))
axes = axes.flatten()  # Flatten to easily iterate over them

# === Create Individual Regression Plots ======================================
for i, country in enumerate(valid_countries):
    ax = axes[i]
    data = df_filtered[df_filtered["Country"] == country]

    # Scatter plot with regression line
    sns.regplot(
        data=data,
        x="Renewable Energy (%)",
        y="Emissions Reduction (%)",
        scatter_kws={'s': 60, 'alpha': 0.7, 'edgecolor': 'w'},
        line_kws={'color': '#e74c3c', 'lw': 2},
        ax=ax
    )

    # Set titles and labels
    ax.set_title(country, fontsize=12, fontweight='bold')
    ax.set_xlabel("Renewable Energy (%)", fontsize=10)
    ax.set_ylabel("Emission Reduction (%)", fontsize=10)
    ax.axhline(0, color='gray', ls='--', alpha=0.5)  # Zero reference line
    ax.axvline(0, color='gray', ls='--', alpha=0.5)

    # Standardize axis ranges (ensuring independence)
    buffer_x, buffer_y = 5, 2  # Axis padding
    ax.set_xlim(data["Renewable Energy (%)"].min() - buffer_x,
                data["Renewable Energy (%)"].max() + buffer_x)
    ax.set_ylim(data["Emissions Reduction (%)"].min() - buffer_y,
                data["Emissions Reduction (%)"].max() + buffer_y)

# Hide any unused subplots if countries < (rows * cols)
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

# === Chart Decoration ========================================================
plt.tight_layout()
plt.suptitle('Clean Energy vs Emission Reduction Correlation\n(Cleaned Data Version)', 
              fontsize=16, y=1.02, fontweight='bold')

# === Output ==================================================================
plt.savefig('renewable_emission_correlation_clean_fixed.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()

# === Statistical Analysis ====================================================
def calculate_correlation(data):
    """Calculate regression statistics"""
    result = stats.linregress(data["Renewable Energy (%)"], 
                             data["Emissions Reduction (%)"])
    return pd.Series({
        'slope': result.slope,
        'intercept': result.intercept,
        'r_value': result.rvalue,
        'p_value': result.pvalue
    })

# Generate correlation results
corr_results = df_filtered.groupby('Country')[["Renewable Energy (%)", "Emissions Reduction (%)"]].apply(calculate_correlation)

print("\nðŸ“Š Correlation Results (Sorted by R-value):")
print(corr_results.sort_values('r_value', ascending=False))

print("\nâœ… Saved cleaned data visualization: renewable_emission_correlation_clean_fixed.png")
```

**Key Findings from the Scatter Plot and Regression Analysis**
1.  A strong positive correlation is observed between renewable energy adoption and emission reduction.
* Most countries exhibit a clear upward trend, indicating that higher renewable energy adoption is strongly associated with greater emission reductions.
2.  The UK, China, and Japan demonstrate the strongest relationships.

## &emsp;4.6  Visualization 6ï¼š
## &emsp;_Scatter Plot with Regression Line:_ 
## &emsp; _Government Investment vs Renewable Energy Share_

**Overview**

This visualization and regression analysis examine the relationship between government investment in renewable energy (Million USD) and the percentage of renewable energy share in total energy consumption across 10 major countries.
* The scatter plot displays individual data points, where:
* X-axis: Government investment in renewable energy (Million USD).
* Y-axis: Renewable energy share (%).
* The regression line (with confidence intervals) represents the trend between these two variables.

The statistical table provides insight into the correlation coefficient (R-value), slope, and significance (P-value).


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats

# --- Data Verification ---
print("\nðŸ” [3.6] Data Validation:")
print("Using cleaned dataset with shape:", df_filled.shape)
print("Key columns missing values:")
print(df_filled[['Government Investment (Million USD)', 'Renewable Energy (%)']].isnull().sum())

# --- Visualization Setup ---
sns.set_theme(style="whitegrid", palette="husl")

# --- Get List of Countries ---
countries = df_filled["Country"].unique()
num_countries = len(countries)

# --- Dynamically Determine Rows & Columns ---
cols = 4  # Keep columns fixed
rows = int(np.ceil(num_countries / cols))  # Adjust rows dynamically

# --- Create Subplots ---
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(16, 4 * rows))  # Adjust height dynamically
axes = axes.flatten()  # Flatten for easy iteration

# --- Store Regression Results ---
regression_results = {}

# --- Plot Each Country Separately ---
for i, country in enumerate(countries):
    ax = axes[i]
    data = df_filled[df_filled["Country"] == country]

    # Scatter plot with regression line
    sns.regplot(
        data=data,
        x="Government Investment (Million USD)",
        y="Renewable Energy (%)",
        scatter_kws={'alpha': 0.7, 'edgecolor': 'w', 's': 60},
        line_kws={'color': '#2c3e50', 'lw': 2.5},
        ax=ax
    )

    # Set titles and labels
    ax.set_title(country, fontsize=12, fontweight='bold')
    ax.set_xlabel("Gov Investment (Million USD)", fontsize=10)
    ax.set_ylabel("Renewable (%)", fontsize=10)
    ax.grid(True, alpha=0.3)

    # --- Perform Regression Analysis ---
    if len(data) > 1:  # Ensure enough data points for regression
        slope, intercept, r_value, p_value, _ = stats.linregress(
            data["Government Investment (Million USD)"],
            data["Renewable Energy (%)"]
        )
        regression_results[country] = [slope, intercept, r_value, p_value]

# --- Hide Unused Subplots ---
for j in range(i + 1, len(axes)):  # Hide remaining empty axes
    axes[j].axis("off")

# --- Adjust Layout & Save ---
plt.tight_layout()
plt.savefig('investment_impact_clean_fixed.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()

print("\nâœ… Saved cleaned data visualization: investment_impact_clean_fixed.png")

# --- Print Regression Results ---
print("\nðŸ“Š Regression Analysis Results (Sorted by R-value):")
regression_df = pd.DataFrame.from_dict(
    regression_results, orient='index', 
    columns=['Slope', 'Intercept', 'R-value', 'P-value']
).sort_values('R-value', ascending=False)

print(regression_df)
```

* Even countries with positive slopes (Brazil, UK, and USA) have low R-values (below 0.35), indicating that government spending alone is not a strong predictor of renewable energy adoption.
